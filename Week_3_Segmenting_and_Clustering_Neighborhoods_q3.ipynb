```python
# Get the content of the wiki page to be processed:
response = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
print("Success" if response.status_code == 200 else "Unexpected status code returned")
```

```python
# Prase the downloaded wiki page:
souped = soup(response.text, 'html.parser')
# Retrieve just the desired section of the content of the wiki page:
postal_codes_table = souped.find("table", class_ = "wikitable")
# Create the data frame:
postal_codes_df = pandas.read_html(str(postal_codes_table))[0]
# Remove those with Borough = "Not assigned"
postal_codes_df = postal_codes_df[postal_codes_df.Borough != "Not assigned"]
# Check for duplicate postal code
duplicated = postal_codes_df.duplicated(subset = "Postal code")
print("No duplicates" if len(duplicated[duplicated == True]) == 0 else "Has duplicate(s)");
```

```python
# Create new dataframe in order to do clustering:
simple_toronto_venues = pd.get_dummies(toronto_venues[["Venue Category"]], prefix="", prefix_sep="")
# add neighborhood column back to dataframe
simple_toronto_venues["Neighborhood"] = toronto_venues["Neighborhood"] 
simple_toronto_venues.head()
```

```python
# One more:
simple_toronto_venues_freq = simple_toronto_venues.groupby("Neighborhood").mean().reset_index()
simple_toronto_venues_freq
```

```python
# Import necessary to do KMeans Clustering:
import sklearn.cluster as c

# Set k = 5
k = 5

# Prepare dataframe:
simple_toronto_venues_freq_clone = simple_toronto_venues_freq.copy()
simple_toronto_venues_clustering = simple_toronto_venues_freq_clone.drop('Neighborhood', 1)
simple_toronto_venues_clustering.head()
```

```python
# Run KMeans
kmeans = c.KMeans(n_clusters = k, random_state=0).fit(simple_toronto_venues_clustering)
# Check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# Add the cluster labels back
simple_toronto_venues_freq.insert(0, 'Cluster Labels', kmeans.labels_)
simple_toronto_venues_freq.head()
```

```python
# Merge back the cluster label
postal_codes_geo_df_clone = postal_codes_geo_df.copy()
postal_codes_geo_df_merged = postal_codes_geo_df_clone.join(simple_toronto_venues_freq.set_index('Neighborhood'), on='Neighborhood')
postal_codes_geo_df_merged.head()
```
