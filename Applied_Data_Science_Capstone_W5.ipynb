```python
# Create a dataframe to store merge of list of neighborhoods and the list of the latitude and longitude of the neighborhoods:
postal_codes_geo_df = postal_codes_df[["Postal code", "Borough", "Neighborhood"]]
# Tidy up indices:
postal_codes_geo_df.reset_index(inplace = True)
postal_codes_geo_df.drop(labels = "index", axis = 1, inplace = True)
# Add latitudes and longitudes to the new dataframe
latitudes = list()
longitudes = list()
for index, row in postal_codes_geo_df.iterrows():
    geo_coord = geo_coordinates_df.loc[row["Postal code"]]
    latitudes.append(geo_coord["Latitude"])
    longitudes.append(geo_coord["Longitude"])
postal_codes_geo_df["Latitude"] = latitudes
postal_codes_geo_df["Longitude"] = longitudes
postal_codes_geo_df.head()
```

```python
# Define a fucntion to call FourSquare:
def get_venues(latitudes, longitudes, category_ids, intent,  limit = 100, radius = 2000):
    url = "https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&v={}&categoryId={}&intent={}&limit={}&ll={}&radius={}"
    url = url.format(client_id, client_secret, v, category_ids, intent, limit, str(latitudes) + "," + str(longitudes), radius)
    response = requests.request("GET", url, headers = headers, data = payload)
    return response.json()

# Define a function to count the numbers of venues of each of the categories from the JSON returned by FourSquare:
def count_categories(json):
    dictionary = {
        "Garden": 0,
        "National Park": 0,
        "Other Great Outdoors": 0,
        "Park": 0,
        "Total Number of Green Zones": 0,
        "Industrial Estate": 0,
        "Power Plant": 0,
        "Waste Facility": 0
    }
    for venue in json["response"]["venues"]:
        for category in venue["categories"]:
            if category["primary"] == True:
                if category["name"] in dictionary:
                    dictionary[category["name"]] += 1 
                if (category["name"] == "Garden" 
                or category["name"] == "National Park" 
                or category["name"] == "Other Great Outdoors" 
                or category["name"] == "Park"):
                    dictionary["Total Number of Green Zones"] +=1
    return dictionary

# Going over the merged dataframe, postal_codes_geo_df, from above to get the count of venues from each of the desired categories:
dataframes = list()
for index, row in postal_codes_geo_df.iterrows():
    json = get_venues(row["Latitude"], row["Longitude"], category_id, intent)
    dictionary = count_categories(json)
    dictionary["Postal code"] = row["Postal code"]
    dataframes.append(pandas.DataFrame(data = dictionary, index = [0]))   
clustering_df = pandas.concat(dataframes)
```

```python
# Drop "Postal code" in order to perform KMeans Clustering:
clustering_df_copy = clustering_df.copy(deep = True)
clustering_df_copy = clustering_df_copy.drop("Postal code", 1)

# Keep "Total Number of Green Zones" and drop "Garden", "National Park", "Other Great Outdoors", "Park" to avoid "noise"
clustering_df_copy = clustering_df_copy.drop("Garden", 1)
clustering_df_copy = clustering_df_copy.drop("National Park", 1)
clustering_df_copy = clustering_df_copy.drop("Other Great Outdoors", 1)
clustering_df_copy = clustering_df_copy.drop("Park", 1)

# Import necessary to do KMeans Clustering:
import sklearn.cluster as clustering

# Set k = 5
k = 5

# Perform KMeans clustering against clustering_df_copy
k_means_result = clustering.KMeans(n_clusters = k, random_state=0).fit(clustering_df_copy)

# Add the cluster label back to the dataframe with contains the list of neighborhoods and postal codes:
postal_codes_geo_df_merged = postal_codes_geo_df.copy(deep = True)
postal_codes_geo_df_merged.insert(0, 'Cluster label', k_means_result.labels_)
# Add the counts back to the dataframe with contains the list of neighborhoods and postal codes:
postal_codes_geo_df_merged = postal_codes_geo_df_merged.join(clustering_df.set_index('Postal code'), on = 'Postal code')

# Verify result dataframe:
postal_codes_geo_df_merged.head()
```
